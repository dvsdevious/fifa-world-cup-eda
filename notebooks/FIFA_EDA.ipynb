
#### 2. Data Reduction

```python
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
```

```python
worldcups_clean = worldcups.copy()
```

```python
X = worldcups_clean[
    ["GoalsScored",
     "QualifiedTeams",
     "MatchesPlayed",
     "Attendance"]
]

X_scaled = StandardScaler().fit_transform(X)

pca = PCA()
pca.fit(X_scaled)

print("\nExplained variance ratio:")
print(pca.explained_variance_ratio_)
```

```python
plt.plot(
    range(1, len(pca.explained_variance_ratio_)+1),
    pca.explained_variance_ratio_.cumsum(),
    marker="o",
)
plt.savefig('../images/pca.png', dpi=300)
```


#### 3. Data Transformation


```python
worldcups_transformed = worldcups_clean.copy()
```

```python
matches_transformed = matches_clean.copy()
```

##### Feature Construction

```python
worldcups_transformed["GoalsPerMatch"] = (
    worldcups_transformed["GoalsScored"] / worldcups_transformed["MatchesPlayed"]
)
```

```python
plt.figure(figsize=(10, 5))
plt.plot(
    worldcups_transformed["Year"],
    worldcups_transformed["GoalsPerMatch"],
    marker="o"
)

plt.title("Average Goals per Match")
plt.xlabel('Year')
plt.ylabel("Goals Per Match")
plt.grid(True)
plt.savefig('../images/average_goals_per_match.png', dpi=300)
plt.show()
```

##### Aggregation

```python
home_goals = matches_transformed[
    ["Home Team Name", "Home Team Goals"]
].rename(
    columns={
        "Home Team Name": "Team",
        "Home Team Goals": "Goals"
    }
)

away_goals = matches_transformed[
    ["Away Team Name", "Away Team Goals"]
].rename(
    columns={
        "Away Team Name": "Team",
        "Away Team Goals": "Goals"
    }
)

all_goals = pd.concat(
    [home_goals, away_goals]
)
```

```python
team_goals = (
    all_goals.groupby("Team")["Goals"].sum().sort_values(ascending=False)
)
```

```python
team_goals.head(10).plot(kind="bar")

plt.title("Top 10 Teams by Total Goals Scored")
plt.ylabel("Goals")
plt.savefig('../images/top_10_teams_by_total_goals_scored.png', dpi=300)
plt.show()
```

```python
avg_attendance = (matches_transformed.groupby("Year")["Attendance"].mean())
```

```python
avg_attendance.head(25).plot(kind="bar")

plt.title("Average Attendance by Year")
plt.ylabel("Year")
plt.savefig('../images/average_attendance_by_year.png', dpi=300)
plt.show()
```

##### Normalization

```python
worldcups_clean[
    ["GoalsScored", "Attendance"]
].head()
```

```python
from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler()

worldcups_transformed[
    ["GoalsScored", "Attendance"]
] = scaler.fit_transform(
    worldcups_transformed[
        ["GoalsScored", "Attendance"]
    ]
)
```

```python
worldcups_transformed[
    ["GoalsScored", "Attendance"]
].head()
```

##### Discretization

```python
worldcups_transformed["AttendanceCategory"] = pd.cut(
    worldcups_transformed["Attendance"],
    bins = 3,
    labels=["Low", "Medium", "High"]
)
```

```python
worldcups_transformed["AttendanceCategory"].value_counts().plot(
    kind="bar"
)

plt.title("Tournament Attendance Categories")
plt.savefig('../images/tournament_attendance_categories.png', dpi=300)
plt.show
```
